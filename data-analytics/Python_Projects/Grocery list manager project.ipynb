```python
# Grocery List Manager

# Menu
def show_menu():
    print(" \n--- Main Menu ---")
    print("[A]dd")
    print("[R]emove")
    print("[V]iew")
    print("[S]lice")
    print("[F]ind")
    print("[C]lear")
    print("[E]xit")

def remove_choice():
    while True:
        mode = input("Remove by [I]index or [N]ame? ").strip().upper()
        if mode in {"I", "N"}:
            return mode
        else:
            print("Invalid choice, Try again.\n")
            continue


# Input Validation
def get_choice():
    while True:
        choose_one = input("Enter your choice: ").strip().upper()
        if choose_one in {"A", "R", "V", "S", "F", "C", "E"}:
            return choose_one
        else:
            print("Invalid choice. Enter your choice again.\n")
            continue
        
# Add
def normalize_item():
    while True:
        item = input("Enter an item: ").strip().lower()
        if len(item) == 0:
            print("Invalid item. Try again.\n")
            continue
        else:
            return item
            
def add_item(groceries, item_name):
    if item_name not in groceries:
        groceries.append(item_name)
        return True
    else:
        return False
    
# View
def display_list(groceries):
        if len(groceries) == 0:
            print("\nNo items yet")
            return
        else:
            print("\n--- Grocery list ---")
            for index, item in enumerate(groceries, start=1):
                print(f"{index}) {item}")

# Remove ( one-function, handles prompting + removal)
def remove_by_index(groceries):
    if len(groceries) == 0:
        print("\nNo items to remove")
        return False

    display_list(groceries)
    while True:
        raw = input("Enter the index of the item to remove (or q to cancel): ").strip()
        if raw.lower() == "q":
            print("\nCancelled.")
            return False
        try:
            index_1based = int(raw)
        except ValueError:
            print("Invalid Index. Try again.\n")
            continue

        index_0 = index_1based -1
        if 0 <= index_0 <len(groceries):
            removed = groceries.pop(index_0)
            print(f"Removed item: {removed}")
            return True
        else:
            print("Index out of range. Try again.")
            continue

def remove_by_name(groceries):
    if len(groceries) == 0:
        print("\nNo items to remove")
        return False

    display_list(groceries)
    while True:
        name = input("Enter the name of the item to remove (or q to cancel):").strip().lower()
        if name == "q":
            print("\nCancelled.")
            return False
        if len(name) == 0:
            print("Enter the name, Try again.\n") 
            continue
        if name in groceries:
            # remove one occurrence
            groceries.remove(name) 
            print(f"Removed item: {name}")
            return True
        else:
            print("Not found. Try again, or enter 'q' to cancel.\n")

# Slice
def display_slice(groceries):
    while True:
        if len(groceries) == 0:
            print("\nNo items yet.")
            return False

        start_1based = input("Enter a start postion (or q to cancel): ").strip()
        if start_1based.lower() == "q":
            print("\nCancelled")
            return False
        if len(start_1based) == 0:
            start_1based =  1
        else:
            try:
                start_1based = int(start_1based)
            except ValueError:
                print("Invalid Index. Try again.\n")
                continue

        end_1based = input("Enter the last position (or q to cancel): ").strip()
        if end_1based.lower() == "q":
            print("\nCancelled")
            return False
        if len(end_1based) == 0:
            end_1based = len(groceries) + 1
        else:
            try:
                end_1based = int(end_1based)
            except ValueError:
                print("Invalid Index. Try again.\n")
                continue

        # Convert to 0-based and clamp to bounds
        start_0 = max(0, min(start_1based - 1, len(groceries)))
        end_0 = max(0, min(end_1based - 1, len(groceries)))

        if start_0 >= end_0:
            print("Start must be less than end.\n")
            continue

        # Display slice with original positions
        print(f"--- Slice {start_0 + 1}..{end_0}(exclusive end) ---")
        for index in range(start_0, end_0):
            print(f"{index+1}) {groceries[index]}")
        return True

# Find/Search
def find_items(groceries):
    if len(groceries) == 0:
        print("\nNo items yet.")
        return False

    while True:
        query = input("Find an item (or q to cancel):").strip().lower()
        if query == "q":
            print("\nCancelled.")
            return False  
        if len(query) == 0:
            print("Invalid query. Try again.\n") 
            continue

        matches = []
        for index, item in enumerate(groceries, start=1):
            if query in item:
                matches.append(index)
        if not matches:
            print("\nNot found")
        else:
            print("\n--- Matches ---")
            for position in matches:
                print(f"{position}) {groceries[position - 1]}")
        return True

# Clear & Polish
def confirm():
    while True:
        prompt = input("This will delete all items. Are you sure? (Y/N): ").strip().lower()
        if prompt in {"y", "yes"}:
            return True
        elif prompt in {"n", "no"}:
            return False
        elif len(prompt) == 0:
            print("Invalid input. Please enter Y or N.\n")
            continue

def clear_list(groceries):
    groceries.clear()
    print("--- list cleared ---\n")
    print(f"List after clearing: {groceries}")

# Skeleton & Single Source of Truth
def main():
    groceries = []
    while True:
        show_menu()
        choice = get_choice()
        if choice == "A":
            item_name = normalize_item()
            added = add_item(groceries, item_name)
            if added:
                print("\nAdded.")
            else:
                print("\nDuplicate item.")
                continue

        elif choice == "V":
            display_list(groceries)
            continue

        elif choice == "R":
            if len(groceries) == 0:
                print("\nNo items to remove")
                continue
            mode = remove_choice()
            if mode == "I":
                remove_by_index(groceries)
            else:
                remove_by_name(groceries)

        elif choice == "S":
            display_slice(groceries)
            continue

        elif choice == "F":
            find_items(groceries)
            continue

        elif choice == "C":
            if len(groceries) == 0:
                print("List is empty. No items to clear.\n")
                continue
            if confirm():
                clear_list(groceries)
            else:
                print("\nCancelled.")
                continue

        elif choice == "E":
            print("Goodbye!\n")
            break

    print("Thanks for using. Goodbye!")


if __name__ == "__main__":
    main()




```
