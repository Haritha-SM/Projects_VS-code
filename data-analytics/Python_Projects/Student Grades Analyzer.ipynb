```python
# Student Grades Analyzer

# Single source of truth for accepted menu tokens
choice = {"A", "AVG", "HG", "LG", "T3", "V", "X"}

# Menu
def show_menu():
    print("\n--- Main menu ---")
    print("[A] → Add (Enter multpiple grades; q to stop)")
    print("[AVG] → Average")
    print("[HG] → Highest Grade ")
    print("[LG] → Lowest Grade")
    print("[T3] → Top Three")
    print("[V] → View")
    print("[X] → Exit")

# Input Validation
def get_choice():
    while True:
        choose_one = input("Enter your choice: ").strip().upper()
        if choose_one in choice:
            return choose_one
        else:
            print("Invalid choice. Enter your choice again.\n")
            continue

# Display
def display_grades(grades):
    if not grades:
        print("\nNo grades yet.")
        return
    # pretty view with count and comma-separated formatting
    formatted = ", ".join(f"{g:.1f}" for g in grades)
    print("\n--- Grades list ---") 
    print(f"Count: {len(grades)}")
    print(formatted)

# Data Entry
def get_valid_grade():
    while True:
        data = input("Enter the grades e.g., 0-100 (or q to quit):").strip()
        if data.lower() == "q":
            print("\nCancelled.\n")
            return None
        if data == "":
            print("Empty input. Try again.\n")
            continue
        try:
            value = float(data)
        except ValueError:
            print("Invalid number. Try again.\n")
            continue
        if 0 <= value <= 100:
            return value
        else: 
            print("Out of range. Enter a value between 0 and 100. \n")
            continue

# Grades list -- multi add flow
def add_grades(grades):
    while True:
        value = get_valid_grade()
        if value is None:
            break
        grades.append(value)
        print("\nGrade recorded.")

# Average grade
def average_grade(grades):
    if not grades:
        return None
    return round(sum(grades) / len(grades), 1)

# Highest grade
def highest_grade(grades):
    if not grades:
        return None
    return max(grades)

# Lowest grade
def lowest_grade(grades):
    if not grades:
        return None
    return min(grades)

# Top three grades
def top_n(grades, n: int = 3):
    if not grades:
        return []
    dup = sorted(grades, reverse=True)
    return dup[ :n]

# Skeleton & Single Source of Truth
def main():
    grades = []
    while True:
        show_menu()
        choice = get_choice()

        if choice == "A":
           add_grades(grades)
           display_grades(grades)

        elif choice == "AVG":
            avg = average_grade(grades)
            if avg is None:
                print("\nNo grades yet.")
            else:
                print(f"\nAverage of grades: {avg:.1f}")
                continue

        elif choice == "HG":
            hi = highest_grade(grades)
            if hi is None:
                print("\nNo grades yet.")
            else:
                print(f"\nHighest grade: {hi:.1f}")
                continue

        elif choice == "LG":
            lo = lowest_grade(grades)
            if lo is None:
                print("\nNo grades yet.")
            else:
                print(f"\nLowest grade: {lo:.1f}")
                continue

        elif choice == "T3":
            top3 = top_n(grades, 3)
            if not top3:
                print("\nNo grades yet.")
            else:
                pretty = ", ".join(f"{g:.1f}" for g in top3)
                print(f"\nTop {min(3, len(grades))} grades: {pretty}")
                continue

        elif choice == "V":
            display_grades(grades)
            continue

        elif choice == "X":
            print("\nGoodbye!\n")
            break

    print("Thanks for using. Goodbye!")


if __name__ == "__main__":
    main()

```
