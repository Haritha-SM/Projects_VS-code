```python
print("Simple Caluculator (type q at any prompt to quit)")

while True:
    # First number
    num1 = input("Enter first number (or) q to quit")
    if num1.lower() == "q":
        break
    try:
        num1 = float(num1)
    except ValueError:
        print("Invalid number.")
        continue

    # Operator
    op = input("Enter the operator +, -, *, / (or) q to quit")
    if op.lower() == "q":
        break
    if op not in {"+", "-", "*", "/"}:
        print("Invalid operator.")
        continue

    # Second number
    num2 = input("Enter second number (or) q to quit")
    if num2 == "q":
        break
    try:
        num2 = float(num2)
    except ValueError:
        print("Invalid number.")
        continue

    # Compute
    if op == "+":
        print(f"The result of {num1} {op} {num2} is: {num1 + num2}")
    elif op == "-":
        print(f"The result of {num1} {op} {num2} is: {num1 - num2}")
    elif op == "*":
        print(f"The result of {num1} {op} {num2} is: {num1 * num2}")
    elif op == "/" and num2 != 0:
        print(f"The result of {num1} {op} {num2} is: {num1 / num2}")    
    else:
        print("Error: Cannot divide by Zero.")
    
print("Thanks for using the caluculator. Goodbye!")
```
