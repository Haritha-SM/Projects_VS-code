```python
# Contact Book

# Contact name
def name():
    # First name (always required)
    first_name = input("Enter first name: ").strip().capitalize()

    # Middle name (optional)
    raw_middle = input("Enter middle name (or 's' to skip middle name): ").strip()
    middle_name = ""    # default
    if raw_middle and raw_middle.lower() != "s":
        middle_name = raw_middle.capitalize()
    
    # Last name (optinal)
    raw_last = input("Enter last name (or 's' to skip last name): ").strip()
    last_name = ""         # default
    if raw_last and raw_last.lower() != "s":
        last_name = raw_last.capitalize()

    # Build final contact name (avoids double spaces)
    parts = [first_name]
    if middle_name:
        parts.append(middle_name)
    if last_name:
        parts.append(last_name)

    contact_name = " ".join(parts)
    print(f"Full Name: {contact_name}")
    return contact_name     # returns the name if you want to use it later


# Contact number
def phone_number():
    # Safe defaults so variables always exist
    prefix = ""
    country_code = ""
    contact_number = ""

    # Prefix: only '+', '-' or skip
    while True:
        raw = input("Enter '+' or '-' (or 's' to skip): ").strip()
        if raw == "" or raw.lower() == "s":
            break
        if raw in {"+", "-"}:
            prefix = raw
            break
        print("Invalid symbol. Enter '+' or '-' (or 's' to skip).")
    
    # Country code: digits only or skip
    while True:
        code = input("Enter country code digits (or 's' to skip): ").strip()
        if code == "" or code.lower() == "s":
            break
        if code.isdigit():
            country_code = code
            break
        print("Country code must be digits only. Try again.")

    # Phone number: digits only, enforce length (example: 10) or skip
    while True:
        num = input("Enter phone number digits (or 's' to skip): ").strip()
        if num == "" or num.lower() == "s":
            break
        if num.isdigit() and len(num) == 10:   # adjust length rule if needed
            contact_number = num
            break
        print("Phone number must be 10 digits. Try again.")

    # Build final phone string (avoids double spaces)
    parts = []
    if country_code and country_code:
        parts.append(prefix + country_code)
    elif country_code:
        parts.append(country_code)
    elif prefix:
        parts.append(prefix)

    if contact_number:
        parts.append(contact_number)
    
    final_number = " ".join(parts)
    if final_number:
        print(f"Phone number: {final_number}")
    else:
        print("No phone number provided.")
    return final_number        # returns the name if you want to use it later

# Add contacts
def add_contact(contacts, name_string, phone_string):
    # Trim spaces
    name = name_string.strip()
    phone = phone_string.strip()

    # Reject empty values
    if not name or not phone:
        return False

    # Check for existing contact (case-insensitive)
    for contact in contacts:
        if contact["name"].lower() == name.lower():
            # Overwrite phone
            contact["phone"] = phone
            return True

    # If not found, add new contact
    contacts.append({"name": name, "phone": phone})
    return True
    


def main():
    contacts = []
    name_string = name()
    phone_string = phone_number()



if __name__ == "__main__":
    main()


```
