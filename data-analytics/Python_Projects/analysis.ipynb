```python
# Even or Odd - robust, modular and user-friendly
def get_number(prompt: str = "Enter a number (or 'q' to quit): "):
    while True:
        raw = input(prompt).strip().lower()
        if raw == "q":
            return None
        try:
            return int(raw)
        except ValueError:
            print("Invalid number. Please enter an integer or 'q' to quit.\n")

def even_odd(number: int):
    if number % 2 == 0:
        return "Even"
    else:
        return "odd"

def display_result(number: int, label: str):
    print(f" The given number {number} is {label}.\n")

def main():
    print("Even or Odd Checker\n")
    while True:
        n = get_number()
        if n is None:
            break
        label = even_odd(n)
        display_result(n, label)


        again = input("Check another number? (Y?N): ").strip().lower()
        if again not in {"y", "yes"}:
            break

    print("\nThanks for using the Even/Odd checker. Goodbye!")

if __name__ == "__main__":
    main()

```
