```python
# Convert Celsius to Fahrenhiet and vis-versa
print("Convert Celsius to Fahrenheit or Fahrenheit to celsius, Choose one")

while True:
      
    # Temperature value
    try:
        temperature = input("Enter the temperature in number:")
        temperature = float(temperature)
    except ValueError:
        print("Invalid number. Try again.\n")
        continue

    # Temperature in degree
    temperature_degree = input("Temperature is in degrees 'C' or 'F':")
    temperature_degree = temperature_degree.upper()
    if temperature_degree not in {'C', 'F'}:
        print("Invalid text. Enter C or F for conversion.\n Try again.\n")
        continue
    
    # Celsius or Fahrenhiet
    convert_temperature = input("Convert temperature into Celsius or Fahrenheit. Enter 'C' for Celsius, or 'F' for Fahrenheit.\n")
    convert_temperature = convert_temperature.upper()
    if convert_temperature not in {"C", "F"}:
        print("Invalid text. Enter C or F for conversion.\n Try again.\n")
        continue
    
    # Check conversion type
    if temperature_degree == convert_temperature:
        print("Entered the same temperature conversion type. Try again.\n")
        continue

    # Compute conversion
    if convert_temperature == "C":
        Celsius = (temperature - 32) * (5/9)
        Celsius = round(Celsius, 1)
        print(f"{temperature}°F converted into: {Celsius}°C")
    elif convert_temperature == "F":
        Fahrenheit = (temperature * 9/5) + 32
        Fahrenheit = round(Fahrenheit, 1)
        print(f"{temperature}°C converted into: {Fahrenheit}°F")

    # Exit the conversion
    exit_loop = input("Enter 'E' to exit, 'R' to run.")
    exit_loop = exit_loop.upper()
    if exit_loop == "E":
        break
    else:
        continue
    

print("Thank you for using temperature conversion, Good bye!")



    

```

```python
# Multiplication table
while True:
    # Enter the multiplication number
    num = input("Enter a number to generate its multiplication table (or q to quit):").strip()
    if num.lower() == "q":
        break
    try:
        num = abs(int(num))
    except ValueError:
        print("Invalid number. Try again. \n")
        continue
    if num == 0:
        print("Please enter a number greater than 0. \n")
        continue

    # Enter the multiplication range (how far to multiply)
    s = input("Enter the maximum multiplier (e.g., 10) (or q to quit):").strip()
    if s.lower() == "q":
        break
    # Default to 10 on blank input
    if s == " ":
        s = "10"
    try:
        s = abs(int(s))
    except ValueError:
        print("Invalid number. Try again. \n")
        continue
    if s < 1:
        print("Please enter a range atleast 1.\n")
        continue

    # Compute multiplication table
    print(f"Multiplication table for {num}")
    for i in range(1, s+1):
        result = num * i
        print(f"{num} x {i} = {result}")
    print()

    # Repeat or exit the loop
    loop = input("Do you want to generate another table? (Y/N) 'Y' for repeat, 'N' for exit").strip()
    if loop.upper() == "Y":
        continue
    elif loop.upper() == "N":
        break
    else:
        print("Unreconized choice. Returning to start.\n ")
        continue

print("Thanks for using the multiplication table!..")


```

```python
def get_int(prompt: str, allow_zero: bool = False, default: int | None = None) -> int | None:
    """
    Prompt for an integer. Returns:
      - the integer if valid,
      - default if user presses Enter and default is provided,
      - None if the user types 'q' (to signal quit).
    """
    while True:
        raw = input(prompt).strip()
        if raw.lower() == "q":
            return None
        if default is not None and raw == "":
            return default
        try:
            val = int(raw)
            if not allow_zero and val == 0:
                print("Please enter a non-zero number.\n")
                continue
            return val
        except ValueError:
            print("Invalid number. Try again.\n")


def ask_yes_no(prompt: str) -> bool | None:
    """
    Ask a yes/no question. Returns:
      - True for yes,
      - False for no,
      - None if the user types 'q'.
    Accepts y/yes/n/no (case-insensitive).
    """
    while True:
        raw = input(prompt).strip().lower()
        if raw == "q":
            return None
        if raw in {"y", "yes"}:
            return True
        if raw in {"n", "no"}:
            return False
        print("Please answer Y or N (or Q to quit).\n")


def print_table(n: int, upto: int) -> None:
    """
    Print a neatly aligned multiplication table for n from 1..upto.
    """
    # Compute column widths for pretty alignment
    max_result = n * upto
    w_num = max(len(str(n)), len("Number"))
    w_i   = max(len(str(upto)), len("x"))
    w_res = max(len(str(max_result)), len("Result"))

    title = f"Multiplication table for {n} (up to × {upto})"
    print("\n" + title)
    print("-" * len(title))

    # Header
    print(f"{'Number':>{w_num}}  {'x':>{w_i}}  {'i':>{w_i}}  =  {'Result':>{w_res}}")
    print(f"{'-'*w_num}  {'-'*w_i}  {'-'*w_i}  -  {'-'*w_res}")

    # Rows
    for i in range(1, upto + 1):
        result = n * i
        print(f"{n:>{w_num}}  {'x':>{w_i}}  {i:>{w_i}}  =  {result:>{w_res}}")
    print()  # spacing


def main() -> None:
    print("Multiplication Table (enter 'q' at any prompt to quit)\n")

    while True:
        # 1) Get the base number
        num = get_int("Enter a number to generate its multiplication table (or 'q' to quit): ")
        if num is None:
            break
        if num < 0:
            # Keep your earlier behavior if you liked abs(); otherwise allow negatives
            num = abs(num)

        # 2) Get the range (default to 10 on blank Enter)
        upto = get_int("Enter the maximum multiplier (default 10): ", allow_zero=False, default=10)
        if upto is None:
            break
        if upto < 1:
            print("Please enter a range of at least 1.\n")
            continue

        # 3) Print table
        print_table(num, upto)

        # 4) Repeat?
        again = ask_yes_no("Generate another table? (Y/N, or Q to quit): ")
        if again is None or again is False:
            break

    print("\nThanks for using the multiplication table!")


if __name__ == "__main__":
    main()
```
